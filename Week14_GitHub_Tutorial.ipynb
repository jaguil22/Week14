## What is Version Control?

Imagine you are writing a research paper. You might save multiple copies as you go:

```
paper_v1.docx
paper_v2.docx
paper_v2_final.docx
paper_v2_final_REAL.docx
paper_v2_final_REAL_revised.docx
```

This is a basic form of version control — but it quickly becomes messy. Which file has the latest changes? What exactly changed between v1 and v2? What if you want to undo a change from three versions ago?

**Version control** solves these problems by:

- Keeping a **complete history** of every change ever made to your files
- Recording **who** made each change, **when**, and **why** (via commit messages)
- Allowing you to **go back** to any previous version at any time
- Letting multiple people work on the same project **without overwriting** each other's work

Instead of saving `paper_v1`, `paper_v2`, etc., you just have **one file** with a timeline of changes underneath it.

## What is Git?

**Git** is a **version control system** — a software tool that tracks changes to files over time. It was created in 2005 by Linus Torvalds (the creator of Linux) to manage the Linux source code.

### Key Concepts in Git

| Concept | What It Means |
|---------|---------------|
| **Repository (repo)** | A project folder tracked by Git — contains your files and their entire change history |
| **Commit** | A snapshot of your files at a specific point in time, like a "save point" in a video game |
| **Commit message** | A short note describing what you changed (e.g., "Added Week 11 homework") |
| **History** | The timeline of all commits — you can view, compare, or revert to any past commit |
| **Branch** | A parallel version of your project (advanced — we won't use this in class) |

### How Git Tracks Changes

Every time you **commit**, Git takes a snapshot of your files and stores it. It does not save the entire file again — it only records **what changed** since the last commit. This makes Git very efficient.

```
Commit 1: "Initial upload of HW Week 11"          ← full file saved
Commit 2: "Fixed map title and added legend"       ← only the changes saved
Commit 3: "Updated analysis with new data"         ← only the changes saved
```

You can always go back to Commit 1 or Commit 2, compare any two commits, or see exactly what lines were added or removed.

### Git is Local

An important point: **Git runs on your own computer.** It does not require the internet. You can track changes, make commits, and browse history entirely offline. But if you want to share your work or back it up online, you need a hosting service — that is where GitHub comes in.

---
## What is GitHub?

**GitHub** is a **cloud platform** that hosts Git repositories online. Think of it this way:

- **Git** = the engine (version control system that tracks changes)
- **GitHub** = the garage (a website where you store and share your Git projects)

### Git vs. GitHub

| | Git | GitHub |
|---|---|---|
| **What is it?** | Software tool (version control) | Website / cloud service |
| **Where does it run?** | On your local computer | On the internet (github.com) |
| **Internet required?** | No | Yes |
| **Purpose** | Track changes to files | Host, share, and collaborate on Git repos |
| **Alternatives** | — (Git is the standard) | GitLab, Bitbucket, Codeberg |
| **Cost** | Free and open-source | Free for public repos; paid plans for advanced features |

### Why Use GitHub?

- **Backup**: Your code is safely stored in the cloud, not just on your laptop
- **Sharing**: Anyone with the link can view your work (if the repo is public)
- **Portfolio**: Your GitHub profile serves as a public portfolio of your technical work — employers look at this
- **Collaboration**: Multiple people can work on the same project with change tracking
- **Integration**: Tools like Google Colab can open notebooks directly from GitHub

## How Version Control Works on GitHub (Web UI)

In this class, we will use GitHub **through the web browser** without installing Git on your computer. Here is how version control works in this workflow:

### The Commit Workflow

Every time you upload, edit, or delete a file on GitHub, you are making a **commit**. Each commit requires a **commit message** — a short description of what you changed.

```
You upload HW_Week11.ipynb
  → GitHub asks for a commit message
  → You write: "Upload HW Week 11"
  → GitHub saves the file + your message + timestamp + your username

Later, you fix a bug and re-upload HW_Week11.ipynb
  → You write: "Fixed choropleth color scale"
  → GitHub saves ONLY what changed (not the whole file again)
```

### Viewing History

You can view the full history of any file:

1. Click on a file in your repository
2. Click **History** (upper right)
3. You see a list of every commit that touched this file
4. Click any commit to see exactly what was added (green) or removed (red)

This is incredibly useful:
- **"My code was working yesterday — what did I change?"** → check the history
- **"I accidentally deleted something important"** → go back to a previous commit
- **"The professor wants to see my progress over time"** → the history shows everything

### Good Commit Messages

Write commit messages that your future self (or your instructor) can understand:

| Bad ❌ | Good ✅ |
|--------|--------|
| `Update` | `Add choropleth map for median income` |
| `asdf` | `Fix broken Census API call` |
| `final version` | `Complete HW Week 12 - all exercises done` |
| `changes` | `Update isochrone function to handle disconnected nodes` |

## Step 1: Create a GitHub Account

2. Click **Sign up** in the upper right corner

## Step 2: Create a Repository

A **repository** (or "repo") is like a folder that holds all of your project files.

1. After signing in, click the **+** icon in the upper right → **New repository**
   - Or go directly to: **https://github.com/new**

2. Fill in the following:

| Field | What to Enter |
|-------|---------------|
| **Repository name** | `geog4500` (or any name you like) |
| **Description** | `Geospatial Programming - Spring 2026` (optional) |
| **Visibility** | Select **Public** (so the instructor can view it) |
| **Initialize** | ✅ Check **Add a README file** |

3. Click **Create repository**

You now have a repository at: `https://github.com/your-username/geog4500`

---
## Step 3: Create Folders to Organize Your Files

GitHub does not have a "create folder" button — instead, you create a folder by including it in a file path.

Let's create the following structure:
```
geog4500/
├── README.md
├── lectures/
├── homework/
└── data/
```

### How to create a folder:

1. In your repository page, click **Add file** → **Create new file**
2. In the filename box at the top, type: `lectures/.gitkeep`
   - When you type the `/`, GitHub automatically creates the folder
   - `.gitkeep` is an empty placeholder file (GitHub cannot store empty folders)
3. Scroll down, keep the default commit message (e.g., "Create .gitkeep"), and click **Commit changes**
4. Repeat for `homework/.gitkeep` and `data/.gitkeep`

Now your repository has three folders!

---
## Step 4: Upload Files

### Uploading a notebook (.ipynb)

1. Navigate to the **folder** where you want to upload (e.g., click on `homework/`)
2. Click **Add file** → **Upload files**
3. Drag and drop your `.ipynb` file (or click "choose your files" to browse)
4. Add a **commit message** describing what you're uploading:
   - Example: `Upload HW Week 11 - Open Source Data`
5. Click **Commit changes**

### Uploading data files (.geojson, .csv, .shp, etc.)

Same process — navigate to `data/`, then upload.

### File size limits

| Limit | Size |
|-------|------|
| Single file (web upload) | 25 MB max |
| Repository total | Recommended under 1 GB |

If your data files are larger than 25 MB, you can either compress them (`.zip`) before uploading, or use Google Drive for large data and GitHub for code only.

## Step 6: Loading Data from Your GitHub Repo in a Notebook

Once your data is on GitHub, you can load it directly in any Colab notebook without downloading:

````python
# Example: loading a GeoJSON file from your GitHub repo
import geopandas as gpd

# Use the "raw" URL (click "Raw" button on any file in GitHub to get this)
# original URL: https://github.com/kim-kyusik/sampledata/blob/main/ATL_bg.geojson
# Replace github.com with raw.githubusercontent.com and remove "blob/"

url = "https://raw.githubusercontent.com/kim-kyusik/sampledata/main/ATL_bg.geojson"
gdf = gpd.read_file(url)
print(f"Loaded {len(gdf)} features")

gdf.head()
````

### How to get the raw URL:

1. Navigate to the file on GitHub
2. Click the **Raw** button (upper right of the file preview)
3. Copy the URL from the browser address bar
4. It will look like: `https://raw.githubusercontent.com/username/repo/main/path/file.ext`

## Step 7: Updating a File (Making Changes)

If you need to update a file you already uploaded:

### For small edits (e.g., README):
1. Click on the file
2. Click the pencil icon (✏️) → edit directly in the browser
3. Click **Commit changes**

### For replacing a notebook:
1. Navigate to the folder containing the file
2. Click **Add file** → **Upload files**
3. Upload the new version of the file (same filename)
4. GitHub will **overwrite** the old version
5. Click **Commit changes**

### Viewing change history:
1. Click on any file → click **History** (upper right)
2. You can see every version of the file and who changed it
3. This is one of the most powerful features of GitHub — you never lose your work!
